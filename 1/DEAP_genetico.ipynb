```python
!pip install deap
import array
import random
import json

import numpy

from deap import algorithms
from deap import base
from deap import creator
from deap import tools
```

# DEAP Genetic algoritms library


```python

with open("d:/dataset/gr24.json", "r") as tsp_data:
    
    tsp = json.load(tsp_data)

distance_map = tsp["DistanceMatrix"]

IND_SIZE = tsp["TourSize"]

#(placeholder)registra un compnente con
#tipo dedato intenger i 
#creano un individuo
creator.create("FitnessMin", base.Fitness, weights=(-1.0,))
creator.create("Individual", array.array, typecode='i', fitness=creator.FitnessMin)

# registrando la funcion
toolbox = base.Toolbox()

#indices con argumentos vector aleatorio creando individuo
toolbox.register("indices", random.sample, range(IND_SIZE), IND_SIZE)

#cea iterable de tipo individuo
#registra la poblacion
toolbox.register("individual", tools.initIterate, creator.Individual, toolbox.indices)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

#funcion deevaluacion suma de todas las distancias
#el ultimo con el primero
#ciclos de suma
#retorna tupla
def evalTSP(individual):
    distance = distance_map[individual[-1]][individual[0]]
    for gene1, gene2 in zip(individual[0:-1], individual[1:]):
        distance += distance_map[gene1][gene2]
    return distance,
#funacion  de registro tools.cxPartialyMatched ej 2 individuos que sevan a cruzar intercambiando
toolbox.register("mate", tools.cxPartialyMatched)
#mutshuffle muta cambiando indices aleatorio seguna la prob de 0.05 
toolbox.register("mutate", tools.mutShuffleIndexes, indpb=0.05)
toolbox.register("select", tools.selTournament, tournsize=30)
toolbox.register("evaluate", evalTSP)

def main():
    #semmilla aleatoria fija
    random.seed(169)
#creacion de poblacion de individuos
    pop = toolbox.population(n=1000)
#el mejor tour
    hof = tools.HallOfFame(1)
    #
    stats = tools.Statistics(lambda ind: ind.fitness.values)
    #
    stats.register("avg", numpy.mean)
    stats.register("std", numpy.std)
    stats.register("min", numpy.min)
    stats.register("max", numpy.max)
    #poblscion tool prob de cruce y mutacion cuantosa correr 
    algorithms.eaSimple(pop, toolbox, 0.7, 0.2, 2000, stats=stats,
                        halloffame=hof)

    return pop, stats, hof

if __name__ == "__main__":
    pop,stats,hof=main()
    print(hof)
    print(evalTSP(hof[0]))
```
